```python
# Khang Tran - khvitran
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# list of top 100 Billboard songs for each year from 2005-2018
top100 = pd.read_csv('songDataTop100ByYearfixtest.csv')

top100['Duration(ms)'] = top100['Duration(ms)']/60000
top100.rename(columns = {'Duration(ms)':'Duration(min)'}, inplace = True) 
#top100
```

```python
# line plot of the data
plt.figure(figsize=(20,13))
singlesDur = sns.lineplot(x = 'Year', y = 'Duration(min)', color = 'purple', data = top100, legend = 'full')

# markers to signify significant user base milestones for Spotify
singlesDur.plot([2008, 2008], [3.333333, 4.4], linewidth = 1, color = 'black')
singlesDur.text(2008, 4.41, 'Spotify launched', horizontalalignment = 'right', fontsize = 14, color = 'black')
singlesDur.plot([2011, 2011], [3.333333, 4.4], linewidth = 1, color = 'black')
singlesDur.text(2011, 4.41, '1 million active users', horizontalalignment = 'right', fontsize = 14, color = 'black')
singlesDur.plot([2012, 2012], [3.333333, 4.4], linewidth = 1, color = 'black')
singlesDur.text(2012, 4.41, '20 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')
singlesDur.plot([2016, 2016], [3.333333, 4.4], linewidth = 1, color = 'black')
singlesDur.text(2016, 4.41, '100 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')

# adjust labels
singlesDur.set_xlabel('Year', fontsize = 20)
singlesDur.set_ylabel('Duration (minutes)', fontsize = 20)
singlesDur.tick_params(labelsize = 14)
singlesDur.set_title('Top 100 Song Durations', fontsize = 22)
singlesDur

# save plots as .jpg
fig = singlesDur.get_figure()
fig.savefig('singlesDurations.jpg')
```

```python
# list of top 200 Billboard albums for each year from 2005-2018
album200 = pd.read_csv("albumDataTop200ByYear.csv")

album200['album_duration(ms)'] = album200['album_duration(ms)']/60000
album200.rename(columns = {'number_of_tracks':'# of tracks',
                          'album_duration(ms)':'Duration(min)'}, inplace = True)
#album200
```

```python
# line plot of the data
plt.figure(figsize=(20,13))
albumCt = sns.lineplot(x = 'Year', y = '# of tracks', color = 'purple', data = album200)

# markers to signify significant user base milestones for Spotify
albumCt.plot([2008, 2008], [13.5, 17], linewidth = 1, color = 'black')
albumCt.text(2008, 17.05, 'Spotify launched', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumCt.plot([2011, 2011], [13.5, 17], linewidth = 1, color = 'black')
albumCt.text(2011, 17.05, '1 million active users', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumCt.plot([2012, 2012], [13.5, 17], linewidth = 1, color = 'black')
albumCt.text(2012, 17.05, '20 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')
albumCt.plot([2016, 2016], [13.5, 17], linewidth = 1, color = 'black')
albumCt.text(2016, 17.05, '100 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')

# adjust labels
albumCt.set_xlabel('Year', fontsize = 20)
albumCt.set_ylabel('Number of tracks', fontsize = 20)
albumCt.tick_params(labelsize = 14)
albumCt.set_title('Top 200 Albums Track Count', fontsize = 22)
albumCt

# save plots as .jpg
fig = albumCt.get_figure()
fig.savefig('albumCt.jpg')
```

```python
# line plot of the data
plt.figure(figsize=(20,13))
albumsDur = sns.lineplot(x = 'Year', y = 'Duration(min)', color = 'purple', data = album200)

# markers to signify significant user base milestones for Spotify
albumsDur.plot([2008, 2008], [52, 64], linewidth = 1, color = 'black')
albumsDur.text(2008, 64.3, 'Spotify launched (2008)', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumsDur.plot([2011, 2011], [52, 64], linewidth = 1, color = 'black')
albumsDur.text(2011, 64.3, '1 million active users', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumsDur.plot([2012, 2012], [52, 64], linewidth = 1, color = 'black')
albumsDur.text(2012, 64.3, '20 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')
albumsDur.plot([2016, 2016], [52, 64], linewidth = 1, color = 'black')
albumsDur.text(2016, 64.3, '100 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')

# adjust labels
albumsDur.set_xlabel('Year', fontsize = 20)
albumsDur.set_ylabel('Duration (minutes)', fontsize = 20)
albumsDur.tick_params(labelsize = 14)
albumsDur.set_title('Top 200 Albums Durations', fontsize = 22)
albumsDur

# save plots as .jpg
fig = albumsDur.get_figure()
fig.savefig('albumsDurations.jpg')
```

```python
# standardize album durations and # of tracks
normAlbum200 = album200.copy()
normAlbum200.rename(columns = {'Duration(min)':'Duration'}, inplace = True)

normAlbum200['Duration'] = ((album200['Duration(min)'] - album200['Duration(min)'].mean())/
                                 (album200['Duration(min)'].std()))
normAlbum200['# of tracks'] = ((album200['# of tracks'] - album200['# of tracks'].mean())/
                               (album200['# of tracks'].std()))
#normAlbum200
```

```python
# line plot to compare track count and album duration
plt.figure(figsize=(20,13))
sns.lineplot(x = 'Year', y = '# of tracks', color = 'red', data = normAlbum200, ci = None)
albumCtDur = sns.lineplot(x = 'Year', y = 'Duration', data = normAlbum200, ci = None)

# markers to signify significant user base milestones for Spotify
albumCtDur.plot([2008, 2008], [-0.15, 0.2], linewidth = 1, color = 'black')
albumCtDur.text(2008, 0.205, 'Spotify launched (2008)', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumCtDur.plot([2011, 2011], [-0.15, 0.2], linewidth = 1, color = 'black')
albumCtDur.text(2011, 0.205, '1 million active users', horizontalalignment = 'right', fontsize = 14, color = 'black')
albumCtDur.plot([2012, 2012], [-0.15, 0.2], linewidth = 1, color = 'black')
albumCtDur.text(2012, 0.205, '20 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')
albumCtDur.plot([2016, 2016], [-0.15, 0.2], linewidth = 1, color = 'black')
albumCtDur.text(2016, 0.205, '100 million active users', horizontalalignment = 'left', fontsize = 14, color = 'black')

# adjust labels
albumCtDur.set_xlabel('Year', fontsize = 20)
albumCtDur.set_ylabel('', fontsize = 20)
albumCtDur.tick_params(labelsize = 14)
albumCtDur.legend(labels = ['Number of tracks', 'Album duration'], fontsize = 16, loc = 'lower left')
albumCtDur.set_title('Track Count vs Album Duration', fontsize = 22)
albumCtDur

# save plots as .jpg
fig = albumCtDur.get_figure()
fig.savefig('albumCtDur.jpg')
```
