```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
top100 = pd.read_csv('songDataTop100ByYearfixtest.csv')

top100['Duration(ms)'] = top100['Duration(ms)']/60000
top100.rename(columns = {'Duration(ms)':'Duration(min)'}, inplace = True) 
#top100
```

```python
singlesDur = sns.lineplot(x = 'Year', y = 'Duration(min)', data = top100, legend = 'full')

singlesDur.plot([2008, 2008], [3.333333, 4.4], linewidth = 1, color = 'green')
singlesDur.text(2008, 4.41, "Spotify Released", horizontalalignment = 'right', size = 'small', color = 'black')

singlesDur.plot([2011, 2011], [3.333333, 4.4], linewidth = 1, color = 'red')
singlesDur.text(2011, 4.41, "Spotify Released (US)", horizontalalignment = 'left', size = 'small', color = 'black')

singlesDur.set_title('Top 100 Songs Durations')
singlesDur
```

```python
album200 = pd.read_csv("albumDataTop200ByYear.csv")

album200['album_duration(ms)'] = album200['album_duration(ms)']/60000
album200.rename(columns = {'number_of_tracks':'# of tracks',
                          'album_duration(ms)':'Duration(min)'}, inplace = True)
#album200
```

```python
albumCt = sns.lineplot(x = 'Year', y = '# of tracks', data = album200)

albumCt.plot([2008, 2008], [13.5, 17], linewidth = 1, color = 'green')
albumCt.text(2008, 17.1, "Spotify Released", horizontalalignment = 'right', size = 'small', color = 'black')

albumCt.plot([2011, 2011], [13.5, 17], linewidth = 1, color = 'red')
albumCt.text(2011, 17.1, "Spotify Released (US)", horizontalalignment = 'left', size = 'small', color = 'black')

albumCt.set_title('Top 200 Albums Track Count')
albumCt
```

```python
albumsDur = sns.lineplot(x = 'Year', y = 'Duration(min)', data = album200)

albumsDur.plot([2008, 2008], [52, 64], linewidth = 1, color = 'green')
albumsDur.text(2008, 64.3, "Spotify Release", horizontalalignment='right', size='small', color='black')

albumsDur.plot([2011, 2011], [52, 64], linewidth=1, color = 'red')
albumsDur.text(2011, 64.3, "Spotify Release (US)", horizontalalignment='left', size='small', color='black')

albumsDur.set_title('Top 200 Albums Durations')
albumsDur
```

```python
normAlbum200 = album200
normAlbum200.rename(columns = {'Duration(min)':'Duration'}, inplace = True)

normAlbum200['Duration'] = ((album200['Duration'] - album200['Duration'].min())/
                                 (album200['Duration'].max() - album200['Duration'].min()))
normAlbum200['# of tracks'] = ((album200['# of tracks'] - album200['# of tracks'].min())/
                               (album200['# of tracks'].max() - album200['# of tracks'].min()))
normAlbum200
```

```python
sns.lineplot(x = 'Year', y = '# of tracks', data = normAlbum200)
albumCtDur = sns.lineplot(x = 'Year', y = 'Duration', data = normAlbum200)
albumCtDur.set(ylabel='')
albumCtDur.legend(labels = ['duration', '# of tracks'])
albumCtDur
```

```python
sns.lineplot(x = 'Year', y = 'value', hue = 'variable', 
             data = pd.melt(normAlbum200, id_vars=['Year'], value_vars=['Duration', '# of tracks']))
```
